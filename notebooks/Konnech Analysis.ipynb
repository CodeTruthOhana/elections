```python
import pandas as pd
```

```python
# Functions

def adjust_frame(dataframe, target_name):
    dataframe = dataframe.drop(['Target Type', 'Metric', 'Database', 'Summary'], axis=1)
    df1 = dataframe[dataframe['Target'] == target_name].reset_index(drop='true')
    df2 = pd.melt(df1, 
        id_vars=['Target'],
        value_vars=dataframe.columns.tolist(),
        var_name='date',
        value_name=target_name)
    df3 = df2.drop(['Target'], axis=1).reset_index(drop=True)
    return df3
```

```python
platform = 'https://raw.githubusercontent.com/'
path = 'CodeTruthOhana/elections/main/data/Konnech/'
file = 'overview-trend-2022-09-12T06_28_31Z.csv'

df = pd.read_csv(platform + path + file)

df.head()
```
